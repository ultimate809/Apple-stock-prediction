```python
import csv
from scipy import stats
import numpy as np
```

```python
date=[]
month=[]
count=[]
i=0
price=[]
```

```python
file=open("/home/shaurya/Downloads/apple.csv",'r')
data=csv.reader(file)
next(data)
for line in data:
    i+=1
    sp=line[0].split("-")
    date.append(int(sp[2]))
    month.append(int(sp[1]))
    count.append(i)
    price.append(float(line[1]))
file.close()
```

```python
slope,intercept,r_value,p_val,std_err=stats.linregress(count,price)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(count,price)
```

```python
def pred(x):
    return slope*x+intercept
```

```python
fitline=[]
for i in count:
    fitline.append(pred(i))
```

```python
plt.scatter(count,price)
plt.scatter(count,fitline)
```

```python
pol=np.poly1d(np.polyfit(count,price,5))
```

```python
plt.scatter(count,price)
plt.scatter(count,pol(count))
```

```python
from sklearn.metrics import r2_score

r2=r2_score(price,pol(count))
r2
```
